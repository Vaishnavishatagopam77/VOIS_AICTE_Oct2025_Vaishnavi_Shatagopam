```python
print("Unique Room Types:")
print(df['room type'].unique())

plt.figure(figsize=(10,6))
sns.countplot(data=df, x='room type', order=df['room type'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Distribution of Room Types")
plt.show()
```

```python
neigh_count = df['neighbourhood group'].value_counts()

print(neigh_count)

plt.figure(figsize=(8,5))
sns.barplot(x=neigh_count.index, y=neigh_count.values)
plt.title("Number of Listings by Neighbourhood Group")
plt.show()
```

```python
avg_price = df.groupby('neighbourhood group')['price'].mean().sort_values(ascending=False)

print(avg_price)

plt.figure(figsize=(8,5))
sns.barplot(x=avg_price.index, y=avg_price.values)
plt.title("Average Price by Neighbourhood Group")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Construction year', y='price', alpha=0.5)
plt.title("Construction Year vs Price")
plt.show()
```

```python
top_hosts = df['host id'].value_counts().head(10)
print(top_hosts)

plt.figure(figsize=(10,6))
sns.barplot(x=top_hosts.index.astype(str), y=top_hosts.values)
plt.xticks(rotation=45)
plt.title("Top 10 Hosts by Number of Listings")
plt.show()
```

```python
verified_reviews = df.groupby('host_identity_verified')['review rate number'].mean()
print(verified_reviews)

sns.boxplot(data=df, x='host_identity_verified', y='review rate number')
plt.title("Verified Hosts vs Review Ratings")
plt.show()
```

```python
corr = df[['price', 'service fee']].corr()
print(corr)

sns.scatterplot(data=df, x='price', y='service fee', alpha=0.5)
plt.title("Price vs Service Fee")
plt.show()
```
